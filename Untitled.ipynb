```python
#procedure
c = [3, 16, 32, 64, 128, 256, 512, 1024, 256, 512, 256, 384, 256]
w = [416, 208, 104, 52, 26, 13, 13, 13, 13, 13, 13, 26, 26]
n = [16, 32, 64, 128, 256, 512, 1024, 256, 512, 255, 128, 256, 255]
L = 0

for i in range(13):
    L += (9 * c[i] + 1) * w[i] * w[i] * n[i] 
    
L += 11 + 1 + 8 + 1
    
t = L / 500000000

print(f'Тактов: {L}')
print(f'Время: {t}')
print(f'FPS: {1/t}')
```

```python
#procedure Mod
L = 0
for i in range(13):
    L += (4 * c[i] + 4) * w[i] * w[i] * n[i]
    
L += 4 * 11 + 4 * 1 + 8 + 1
    
t = L / 500000000

print(f'Тактов: {L}')
print(f'Время: {t}')
print(f'FPS: {1/t}')
```
